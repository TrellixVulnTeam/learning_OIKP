```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### HEATMAP

```python
flights = sns.load_dataset('flights')

#data  heatmap? 
#data    heatmap?
#    .     

plt.figure(figsize=(12,5))
sns.heatmap(flights.pivot_table(index='month',columns='year',values='passengers'),annot=True,fmt='d')
plt.show()
```

#### Box plot

```python
#  .
np.random.seed(333)
ri = np.random.randint(1,101,100)
df = pd.DataFrame(ri,columns=['num'])

plt.figure(figsize=(10,5))
df.boxplot(column=['num'])
plt.yticks(np.arange(0,101,step=10),rotation=20) #! 
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
#1. total_bill   tip  50%  tip .
sns.boxplot(x=tips['total_bill'])
plt.show()

#x tips    box-and-whiskers plot.
```

```python
#2.  total_bill   box plot,  smoker
plt.figure(figsize=(10,8))
sns.boxplot(y=tips['total_bill'],x=tips['day'],hue=tips['smoker'])
plt.show()
```

```python
#3.  total_bill   box plot  sex
plt.figure(figsize=(10,8))
sns.boxplot(y=tips['total_bill'],x=tips['day'],hue=tips['sex'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(y=tips['tip'],x=tips['size'])
plt.show()
```

```python
tips.boxplot(column='tip',by='size',figsize=(10,8))

plt.show()
```

#### Lmplot
- to validate linear model

```python
iris = sns.load_dataset('iris')
#sepal length petal length  ?
plt.figure(figsize=(12,4))
sns.lmplot(x='sepal_length',y='petal_length',data=iris[iris['species']=='setosa'])
sns.lmplot(x='sepal_width',y='petal_width',data=iris[iris['species']=='virginica'])
sns.lmplot(x='sepal_width',y='petal_width',data=iris[iris['species']=='versicolor'])

#or
sns.lmplot(x='sepal_width',y='petal_width',data=iris,hue='species') #by putting hue, you can legending and visualize them better.


```

```python
sns.lmplot(x='sepal_width',y='petal_length',data=iris,hue='species')
```
